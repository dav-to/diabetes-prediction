# Diabetes Classification

## 1. About 

In this notebook prediction of diabetes for patients is explored. A single decision tree and a random forest model is used.
The data set used is from Islam et al. (2020) and contains information collected by questionnaires to patients. 


## 2.  Data Exploration 

Before bulding the classification models we will load the data and explore it.

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
from sklearn.model_selection import train_test_split, GridSearchCV
import matplotlib.pyplot as plt
from sklearn import metrics 
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
```

```python
# loading data
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00529/diabetes_data_upload.csv')
data.head()
```

```python
data.shape
```

There are no missing values in the dataset:

```python
data.isnull().sum()
```

Let's look at the class distribution in the data:

```python
data.groupby('class').size()
```

We have two different classes, that is posetive (having diabetes) and negative (not having diabetes). The number of observation differs but a ratio of 16:10 is accepted here, without other meassures such as for example oversampling of the minority class.

We now split the datan into training and testing using 30% for testing: 

```python
train, test = train_test_split(data, test_size =0.3, stratify=data['class'], random_state = 123)
```

## 3. Classification

We will now try to predict if a person has diabetes or not, starting with a single decision tree. However, first we transform the data so that we can use our categorical features when training the tree with scikit-learn. We do this by applying one-hot encoding to the categorical features. We define a function below to use for the transformation of the training and test data.

```python
def oneHotData(dataOrg):
    dataOrg = dataOrg.reset_index()
    enc = OneHotEncoder(sparse=False)
    enc_data = enc.fit_transform(dataOrg[['Gender', 'Polyuria', 'Polydipsia',
    'sudden weight loss', 'weakness', 'Polyphagia', 'Genital thrush',
    'visual blurring', 'Itching', 'Irritability', 'delayed healing',
    'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity']])
     # getting the names 
    columnNames=enc.get_feature_names(['Gender', 'Polyuria', 'Polydipsia',
    'sudden weight loss', 'weakness', 'Polyphagia', 'Genital thrush',
    'visual blurring', 'Itching', 'Irritability', 'delayed healing',
    'partial paresis', 'muscle stiffness', 'Alopecia', 'Obesity'])
    # make data frame and concatenate class and label
    dataEncoded = pd.DataFrame(data=enc_data, columns=columnNames)
    dataEncoded = pd.concat([dataEncoded, dataOrg[['Age', 'class']]], axis = 1)
    
    return dataEncoded
```

```python
# training data
trainEnc=oneHotData(train)
# test data
testEnc=oneHotData(test)
```

We will now start with trying to classify patients as having or not having diabetes using a single decision tree.

```python
# first split the training and testing data into labels and features 
x_train = trainEnc.drop(columns=['class'])
y_train = trainEnc['class']
x_test = testEnc.drop(columns=['class'])
y_test = testEnc['class']
```

```python
dTree = DecisionTreeClassifier(max_depth=4, random_state=123)
dTree.fit(x_train, y_train)
```

Testing the tree on the test set:

```python
# predictions for the test set 
predictionDt = dTree.predict(x_test)
accuracyDt = metrics.accuracy_score(predictionDt, y_test)
print(f'The test accuracy of the decision tree is: {accuracyDt}')
```

The model seems to preform pretty well with an accuracy of close to 90%, even if we are using a single tree and a max depth of 4. The advantage of this simple model is that we can plot the tree and see its decision rule:

```python
colNames = list(x_train.columns)
plt.figure(figsize=(20,20))
plot_tree(dTree, feature_names = colNames, class_names = ['Negative', 'Positive'], filled=True, fontsize=10)
plt.show()
```

The plot of the tree confirms the importance of the symptoms polyuria and polydipsia for detecting diabetes. 

We will know test a random forest model. For tuning of the hyperparameters; number of trees, number of features to consider for split and the max depth we use 5-fold cross validation and grid search. 

```python
# defining the grid we consider
grid = {
    'max_features':[5,8,10,15,20],
    'n_estimators':[100,200,300,400,500],
    'max_depth':[1,2,4,6,8,10]
}

# defining model
randomForestModel = RandomForestClassifier(random_state = 123)
# grid search and CV setup
gridSearch = GridSearchCV(estimator = randomForestModel, param_grid = grid, cv = 5, n_jobs=-1, verbose = 2)
```

```python
# running CV
gridSearch.fit(x_train, y_train)
```

We select the model with the following hyperparameters from the grid search: 

```python
gridSearch.best_params_
```

Finally, we test the selected model on the test set. The model perform pretty well with an accuracy above 95%. 

```python
rfBest = gridSearch.best_estimator_
# predict for test set 
predictionRf = rfBest.predict(x_test)
accuracyRf = metrics.accuracy_score(predictionRf, y_test)
print(f'The test accuracy of the random forest model is: {accuracyRf}')
```

We also look at the confusion matrix:

```python
metrics.plot_confusion_matrix(rfBest, x_test, y_test, display_labels =['Negative', 'Positive'], cmap=plt.cm.Blues, normalize=None)
```

For the sensitivity and specificity we have:

```python
print(f'Sensitivity is {91/96} and specificity is {58/60}')
```

The sensitivity and specificity are close, but sensitivity slightly lower. In this case we might be more concerned with the model being able to correctly classify a patient with diabetes and not missing some (false negative). Additional work could be to adjust for the lower sensitivity with a possible comprise of a higher specificity. But overall, the model is quite good. By letting a patient answer a questionnaire about symptoms and other factors we can with a high accuracy of around 96 % predict if the patient has diabetes or not.
